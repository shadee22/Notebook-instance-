```python
def load_data(csv_file, year):
    import pandas as pd
    # Load whole data set and filtering calender year data
    df = pd.read_csv(csv_file)
    # Convert the 'Conversion Time' column to datetime
    df['Conversion Time'] = pd.to_datetime(df['Conversion Time'], format='%d/%m/%Y %H:%M')
    
    # Filter the rows where the year is the input year and the Conversion Time is before the current date
    df = df[(df['Conversion Time'].dt.year == year) ]
    return df

df = load_data('m1 latest/devnodes.csv', 2024)
```

```python
from dateutil import parser
from datetime import datetime

# Assuming df is your DataFrame and 'Conversion Time' is your column with dates
dates = df['Conversion Time']

def preprocess_dates(date_list):
    processed_dates = []
    for date in date_list:
        try:
            # Convert date to string
            date_str = str(date)
            # Try to parse the date assuming day-first format
            date_obj = parser.parse(date_str, dayfirst=True)
            # Reformat to the desired format
            correct_format = date_obj.strftime("%d-%m-%Y")
            processed_dates.append(correct_format)
        except ValueError:
            # Handle the error if parsing fails
            print(f"Error parsing date: {date_str}")
            processed_dates.append(date_str)
    return processed_dates

# Preprocess the dates
processed_dates = preprocess_dates(dates)

# Add the processed dates back to the DataFrame
df['Processed Conversion Time'] = processed_dates

# Print the DataFrame with the processed dates
print(df[['Conversion Time', 'Processed Conversion Time']])
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

# Load your data into the 'data' DataFrame
# data = pd.read_csv('your_data_file.csv')  # Example of loading data

# Convert the 'Conversion Time' column to datetime format
data['Conversion Time'] = pd.to_datetime(data['Conversion Time'])

# Define the current time
current_time = datetime.now()

# Create a dataframe to track user activity
user_activity = data.groupby('EUID')['Conversion Time'].agg(['min', 'max', 'count']).reset_index()
user_activity.columns = ['EUID', 'First Conversion Time', 'Last Conversion Time', 'Frequency']

# Calculate Recency (in days)
user_activity['Recency'] = (current_time - user_activity['Last Conversion Time']).dt.days

# Define the categorization logic
def categorize_user(row, current_time):
    if row['Frequency'] == 1 and row['Recency'] <= 60:
        return 'A'  # New User
    elif row['Recency'] <= 60:
        return 'B'  # Active User
    elif row['Recency'] > 120:
        return 'D'  # Not Active for Past 120 Days
    elif row['Recency'] > 60:
        return 'C'  # Not Active for Past 60 Days
    else:
        return 'Other'

# Apply the categorization
user_activity['Label'] = user_activity.apply(categorize_user, axis=1, current_time=current_time)

# user_labels = user_activity[['EUID', 'Last Conversion Time', 'Frequency', 'Recency',  'Label']]
user_labels = user_activity[['EUID', 'Last Conversion Time',  'Label']]

# Display the updated dataframe with EUID and labels
user_labels.head()
```
