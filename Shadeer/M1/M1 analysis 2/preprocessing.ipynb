## Loading data

```python
import pandas as pd

# Load your dataset
df = pd.read_csv('./datasource/kavinash_req.csv')

df.head()
```

## Preprocessing

```python
df['Advertiser'] = df['Advertiser'].replace('Shopee Singapore', 'Shopee SG').replace('Shopee SG - CPS', 'Shopee SG').replace('Trip.com', 'tripcom').replace('Decathlon', 'Decathlon SG')
```

```python
df.head()
```

### Only using required Cols

```python
required_cols = ['Conversion ID', 'Advertiser', 'Status', 'Conversion Time','Estimate M1 Cashback Value' ]
df = df[required_cols]
```

### Handling null values

```python
df.isnull().sum()
```

```python

df = df.dropna()
```

## Saving Processed dataset

```python
df.to_csv('./datasource/preprocessed_data.csv', index=False)
```
