```python
import matplotlib.pyplot as plt
k = df.Cluster.nunique()
# Iterate through each cluster
for cluster_id in range(k):
    # Filter the DataFrame to include only data points belonging to the current cluster
    cluster_data = df[df['Cluster'] == cluster_id]
    
    print(f"Cluster {cluster_id} Insights:")
    
    # Summary statistics
    print("Summary Statistics:")
    print(cluster_data.describe())
    
    # Visualization of features
    plt.figure(figsize=(12, 6))
    for i, feature in enumerate(features):
        plt.subplot(2, 3, i+1)
        plt.hist(cluster_data[feature], bins=20, alpha=0.7)
        plt.title(feature)
    plt.tight_layout()
    plt.show()
```

## Final With Labels for Cluster


```python
import pandas as pd
import numpy as np

merged_data = pd.read_csv("./DataSources/merged_data_with_cluster_and_label.csv")

# Assuming merged_data is your DataFrame and 'Cluster' is the column containing cluster information and 'UID' is the column containing UID information
cluster_dict = {}

for cluster, uid in merged_data[['Cluster', 'UID']].values:
    if cluster not in cluster_dict:
        cluster_dict[cluster] = {'label': '', 'uid_list': []}  # Rearranged the order of keys
    cluster_dict[cluster]['uid_list'].append(uid)

# Adding labels to clusters
cluster_dict[0]['label'] = "high frequency buyers"
cluster_dict[1]['label'] = "low frequency buyers"
cluster_dict[2]['label'] = "Champions"
cluster_dict[3]['label'] = "Value Seekers"
# Add labels for other clusters as needed

# Optionally, if you want to convert the UID lists to sets to remove duplicates
for cluster, data in cluster_dict.items():
    data['uid_list'] = list(set(data['uid_list']))
    
cluster_dict
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv("./DataSources/merged_data_with_cluster_and_label.csv")
# Calculate the threshold for the top 25% frequency score
frequency_threshold = data['Frequency_Score'].quantile(0.75)

# Identify high frequency buyers based on the top 25% frequency score
high_frequency_buyers = data[data['Frequency_Score'] >= frequency_threshold]

# Display the number of high frequency buyers and a few sample rows from this filtered data
high_frequency_buyers_count = high_frequency_buyers.shape[0]
# high_frequency_buyers.sample(5), high_frequency_buyers_count

hfb_user_ids = high_frequency_buyers.UID
```

```python
# Calculate the threshold for the top 25% based on 'Total Spend by Customer'
total_spend_threshold = data['Total Spend by Customer'].quantile(0.75)

# Identify high value buyers based on the top 25% of total spend by customer
high_value_buyers_based_on_spend = data[data['Total Spend by Customer'] >= total_spend_threshold]

# Display the number of high value buyers and a few sample rows from this filtered data
high_value_buyers_based_on_spend_count = high_value_buyers_based_on_spend.shape[0]
high_value_buyers_based_on_spend.sample(5), high_value_buyers_based_on_spend_count
```
