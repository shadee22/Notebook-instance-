```python
import pandas as pd
df = pd.read_csv('../VOX_Data_PERCENTAGE_With_Seg_Counts.csv')
```

```python
import matplotlib.pyplot as plt

counts = df["offer_id"].value_counts()
plt.figure(figsize=(10, 6))
counts.plot(kind='bar')
plt.title('Offer ID Counts')
plt.xlabel('Offer ID')
plt.ylabel('Offer Title')
plt.show()
```

```python

import matplotlib.pyplot as plt

counts = df["offer_title"].value_counts()
plt.figure(figsize=(10, 6))
counts.plot(kind='bar')
plt.title('Offer Title Counts')
plt.xlabel('Offer Title')
plt.ylabel('Count')
plt.show()
```
